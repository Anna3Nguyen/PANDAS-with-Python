```python
# 1. importing libraries
import pandas as pd
```

```python
# 2. importing data
iris = pd.read_csv('iris.csv')
```

```python
# 3. Understand what this data looks like
print(iris.head()) # gives very first few lines
print(iris.tail()) # gives very last 5 lines
print(iris.shape)
```

```python
iris.sepal_length[0:2]
```

```python
# preliminary statistical analysis to this
# mean, median, minimum, maximum
print(iris.sepal_length.min()) # finding the minimum
print(iris.sepal_length.max()) # maximum
print(iris.sepal_length.mean())
```

```python
print(iris.petal_length.min())
print(iris.petal_length.max())
print(iris.petal_length.mean())
```

```python
# plotting data
iris.sepal_length.hist()
```

```python
# general code for comparing histogram data
iris[['sepal_width','sepal_length']].plot.hist(alpha =0.9);
```
